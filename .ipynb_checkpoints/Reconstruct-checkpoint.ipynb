```python
# Calibration for Stereo Imaging devices
calibration = pickle.load(open('C:\\Users\\brije\\Documents\\reconstruct\\calibration.pickle', 'rb'))
fx = calibration['fx']
fy = calibration['fy']
cx = calibration['cx']
cy = calibration['cy']
dist = calibration['dist']

fl = (fx + fy) / 2
offset = np.array([[cx, cy]]).T
camL = Camera(f = fl, c = offset, t = np.array([[0, 0, 0]]).T, R = makerotation(0, 0, 0))
camR = Camera(f = fl, c = offset, t = np.array([[0, 0, 0]]).T, R = makerotation(0, 0, 0))

imgL = plt.imread('C:\\Users\\brije\\Documents\\reconstruct\\calib_jpg_u\\frame_C1_01.jpg')
ret, cornersL = cv2.findChessboardCorners(imgL, (8, 6), None)
pts2L = cornersL.squeeze().T

imgR = plt.imread('C:\\Users\\brije\\Documents\\reconstruct\\calib_jpg_u\\frame_C0_01.jpg')
ret, cornersR = cv2.findChessboardCorners(imgR, (8, 6), None)
pts2R = cornersR.squeeze().T

pts3 = np.zeros((3, 6 * 8))
yy, xx = np.meshgrid(np.arange(8), np.arange(6))
pts3[0,:] = 2.8 * xx.reshape(1, -1)
pts3[1,:] = 2.8 * yy.reshape(1, -1)

camLParams = [0, -30, 0, 0, 0, -15]
camRParams = [0, 30, 0, 0, 0, -15]
camL = calibratePose(pts3, pts2L, camL, camLParams)
camR = calibratePose(pts3, pts2R, camR, camRParams)

plt.figure()
pts2Lp = camL.project(pts3)
plt.imshow(imgL)
plt.plot(pts2Lp[0,:],pts2Lp[1,:],'bx')
plt.plot(pts2L[0,:],pts2L[1,:],'ro')
plt.show()

plt.figure()
pts2Rp = camR.project(pts3)
plt.imshow(imgR)
plt.plot(pts2Rp[0,:],pts2Rp[1,:],'bx')
plt.plot(pts2R[0,:],pts2R[1,:],'ro')
plt.show()
```

```python
def generateMesh(triThresh, bound, pts2L, pts2R, pts3, color):
    
    # Bounding box pruning
    goodPts = np.nonzero((pts3[0,:] > bound[0]) & (pts3[0,:] < bound[1]) &
                        (pts3[1,:] > bound[2]) & (pts3[1,:] < bound[3]) & (pts3[2,:] > bound[4]) & (pts3[2,:] < bound[5]))
    pts3 = pts3[:,goodPts[0]]
    pts2L = pts2L[:,goodPts[0]]
    pts2R = pts2R[:,goodPts[0]]
    color = color[:,goodPts[0]]
    
    # Initial Triangulation
    Triangles =  scipy.spatial.Delaunay(pts2L.T)
    tri = Triangles.simplices
      
    # Neighbor Smoothing
    def neighbors(pindex, triang):
        return triang.vertex_neighbor_vertices[1][triang.vertex_neighbor_vertices[0][pindex]:triang.vertex_neighbor_vertices[0][pindex+1]]
    for x in range (pts3.shape[1]):
        pts3[:,x] = np.mean(pts3[:,neighbors(x,Triangles)],axis=1)
    for x in range (pts3.shape[1]):
        pts3[:,x] = np.mean(pts3[:,neighbors(x,Triangles)],axis=1) 
    for x in range (pts3.shape[1]):
        pts3[:,x] = np.mean(pts3[:,neighbors(x,Triangles)],axis=1)
        
    # Creating triangles
    a1 = np.sqrt(np.sum(np.power(pts3[:,tri[:,0]] - pts3[:,tri[:,1]], 2), axis = 0))
    a2 = np.sqrt(np.sum(np.power(pts3[:,tri[:,0]] - pts3[:,tri[:,2]], 2), axis = 0))
    a3 = np.sqrt(np.sum(np.power(pts3[:,tri[:,1]] - pts3[:,tri[:,2]], 2), axis = 0))
    
    # Thresholding triangles
    goodTri = (a1 < triThresh) & (a2 < triThresh) & (a3 < triThresh)
    tri = tri[goodTri,:]

    # Mesh Smoothing
    tokeep=np.unique(tri)
    map = np.zeros(pts3.shape[1])
    pts3=pts3[:,tokeep]
    color = color[:,tokeep]

    # Removing points not referenced in triangle
    map[tokeep] = np.arange(0,tokeep.shape[0])
    tri=map[tri]

    return pts2L, pts2R, pts3, color, tri
```
